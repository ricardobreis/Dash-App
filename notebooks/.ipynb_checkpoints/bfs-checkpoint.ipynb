```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(3, 5)
```

```python
nx.draw(G, with_labels=True)
```

## BFS

```python
list(nx.bfs_edges(G, 3))
```

```python
bfs_edges = nx.bfs_edges(G, 3)
```

```python
E = nx.DiGraph(bfs_edges)
```

```python
nx.draw(E, with_labels=True)
```

## BFS Reverso

```python
F = nx.DiGraph()
```

```python
list(nx.bfs_edges(G, 3, reverse=True))
```

```python
bfs_reverse_edges = nx.bfs_edges(G, 3, reverse=True)
```

```python
for i, j in bfs_reverse_edges:
    F.add_edge(j, i)
```

```python
nx.draw(F, with_labels=True)
```

```python
plt.figure(figsize=(10,4))

ax1 = plt.subplot(1,2,1)
ax1.set_title("Descendentes")
nx.draw(E, with_labels=True, ax=ax1, node_color=["red", "#1f78b4", "#1f78b4"])

ax2 = plt.subplot(1,2,2)
ax2.set_title("Ancestrais")
nx.draw(F, with_labels=True, ax=ax2, node_color=["#1f78b4", "red", "#1f78b4"])

plt.show()
```

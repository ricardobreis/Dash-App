```python
import pandas as pd
import networkx as nx
import matplotlib as plt
```

```python
G = nx.DiGraph()
G.add_edge(1, 2)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(3, 5)
```

```python
nx.draw(G, with_labels=True)
```

## BFS

```python
list(nx.bfs_edges(G, 3))
```

```python
bfs_edges = nx.bfs_edges(G, 3)
```

```python
E = nx.DiGraph(bfs_edges)
```

```python
nx.draw(E, with_labels=True)
```

## BFS Reverso

```python
F = nx.DiGraph()
```

```python
list(nx.bfs_edges(G, 3, reverse=True))
```

```python
bfs_reverse_edges = nx.bfs_edges(G, 3, reverse=True)
```

```python
for i, j in bfs_reverse_edges:
    F.add_edge(j, i)
```

```python
nx.draw(F, with_labels=True)
```
